```python
# 常用的统计计算
# sum,max,min,mean(值的平均数)
# axis=0 按列统计，axis=1按行统计
# skipna 排除缺失值， 默认为True
import numpy as np
import pandas as pd

df_obj = pd.DataFrame(np.random.randn(5,4), columns = ['a', 'b', 'c', 'd'])
print(df_obj)

print(df_obj.sum())
print("-"*50)
print(df_obj.max())
print("-"*50)
print(df_obj.min(axis=1, skipna=False))
print("-"*50)
```
