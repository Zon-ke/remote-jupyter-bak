```python
# transpose() 数组转置
arr = np.random.rand(9,3)    # 2x3 数组
print(arr)  
print("-"*50)
print(arr.transpose()) # 转换为 3x2 数组
print("-"*50)
df_obj = pd.DataFrame(arr, columns=["y_0", "y_1","y_2"], index = range(1,10))
print(df_obj)
print("-"*50)
x = df_obj.index
print(type(df_obj.index))


y_0 = list(df_obj["y_0"])
# print(y_0)
y_1 = list(df_obj["y_1"])
y_2 = list(df_obj["y_2"])

# plt.figure(figsize=(25,12), dpi=120)
plt.plot(x, y_0, label="y_0")
plt.plot(x, y_1, label="y_1")
plt.plot(x, y_2, label="y_2")
_x = list(x)
_xticks_labels = [i for i in _x]

plt.legend(loc="upper right")
plt.show()
print("-"*50)
```
